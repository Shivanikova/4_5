## Предобработка файлов

```python
# Загрузка файлов
customers_df = pd.read_csv('customers.csv')
organizations_df = pd.read_csv('organizations.csv')
people_df = pd.read_csv('people.csv')
```

### Предобработка файла "customers.csv"

```python
customers = customers_df.copy()
```

```python
customers
```

```python
customers.info()
```

```python
# преобразование типа данных столбца 'Subscription Date'
customers['Subscription Date'] = customers['Subscription Date'].astype('datetime64[ns]')
```

```python
customers.info()
```

```python
# добавление столбца с номером года, в который была совершена подписка
customers['Subscription Year'] = customers['Subscription Date'].dt.to_period('Y')
```

```python
customers.info()
```

```python
customers.head(2)
```

```python
# Добавление столбца 'Groups' с разделением файла на 10 груп
customers['Groups'] = pd.qcut(customers['Index'], q=10, labels=['01','02','03','04','05','06','07','08','09','10'])
```

```python
customers.info()
```

```python
customers.head(2)
```

```python
customers.tail(2)
```

```python
customers['Groups'].unique()
```

```python
customers.to_csv('customers_with_groups.csv', index = False, sep = '\t')
```
